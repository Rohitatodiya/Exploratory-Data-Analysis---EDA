# **Problem Statement**

Suppose you are working in the data-driven domain at Airbnb. You have a dataset and need to derive insight from it to answer key business questions, as company officials aim to grow the business.



1. What is the distribution of listing prices?
2. How are different room types distributed?
3. How are listing distributed across different neighborhoods?
4. What is the relationship between price and room type?
5. how has the number of reviews changed over time?

**Dataset:-** /content/Airbnb_Open_Data.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Airbnb_Open_Data.csv')
```

```python
df.head()
```

```python
df.columns
```

# Cheacking Missing Values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling Missing Value

```python
df['last review'] = pd.to_datetime(df['last review'],errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0,'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset =["NAME","host name"], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns = ['license','house_rules'], errors = 'ignore')
```

```python
df.head()
```

```python
# Remove dollar signs and convert to float
df['price'] = df['price'].replace('[\$,]','', regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','', regex = True).astype(float)
```

```python
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# Descriptive Statistics

```python
df.describe()
```

# Visualization

# Que 1. What is the distribution of listing prices?

```python
plt.figure(figsize =(10,6))
sns.histplot(df['price'], bins =50, kde = True, color = 'red')
plt.title('Distribution of Listing Price')
plt.xlabel("Price $")
plt.ylabel('Frequency')
plt.show()
```

The Histogram show a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listing in any specific price range. The KDE line helps visualize this even spread more clearly, confirming that the dataset contains listing with a wide variety of prices.

# Que 2. How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x = 'room type', data = df, color = 'hotpink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

# Que 3. How are listing distributed across different neighborhoods?

```python
plt.figure(figsize=(12,8))
sns.countplot(y = 'neighbourhood group', data = df, color ='lightgreen', order= df['neighbourhood group'].value_counts().index)
plt.title('Number of Listing by Neighbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

# Que 4. What is the relationship between price and room type?

```python
plt.figure(figsize = (10,6))
sns.boxplot(x = 'room type', y = 'price', hue = 'room type', data = df, palette = 'Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title = 'Room Type')
plt.show()
```

# Que 5. How has the number of reviews changed over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize =(12,6))
reviews_over_time.plot(kind='line', color = 'red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
